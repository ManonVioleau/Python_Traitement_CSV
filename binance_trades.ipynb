```python
import pandas
from functions import get_value_in_USD, get_value_in_USDT_width_binance

data = pandas.read_csv(r'/home/manon/Bureau/Adrien/Adrien2/csv/binance_new/binance_trades_all_2021-09-12.csv')
print(data.head(5))
```

Mise en forme

```python
new_datas = {'Portfolio' : [], 'Type' : [], 'Date' : [], 'Received Currency' : [], 'Received Quantity' : [], 'Sent Currency' : [], 'Sent Quantity' : [], 'Fee Currency' : [], 'Fee Quantity' : [], 'Trade value in USDT' : [], 'Fees value in USDT' : [], 'Total trade value in USDT' : [], 'Timestamp' : []}

for index, row in data.iterrows():
    new_datas['Portfolio'].append('Binance SPOT')
    new_datas['Date'].append(row['Date(UTC)'])        
    new_datas['Timestamp'].append((pandas.to_datetime(row['Date(UTC)']) - pandas.Timestamp("1970-01-01 00:00:00")) // pandas.Timedelta('1ms'))
    new_datas['Type'].append('Trade')
    received_cur = ''
    sent_cur = ''
    received_quant = 0.0
    sent_quant = 0.0
    if len(str(row['Pair']).split('USDT')) > 1 and str(row['Pair']).split('USDT')[1] == '':
        if row['Side'] == 'SELL':
            received_cur += str(row['Pair']).split('USDT')[0]
            sent_cur += 'USDT'
        else:
            sent_cur += str(row['Pair']).split('USDT')[0]
            received_cur += 'USDT'
    elif len(str(row['Pair']).split('ETH')) > 1 and str(row['Pair']).split('ETH')[1] == '':
        if row['Side'] == 'SELL':
            received_cur += str(row['Pair']).split('ETH')[0]
            sent_cur += 'ETH'
        else:
            sent_cur += str(row['Pair']).split('ETH')[0]
            received_cur += 'ETH'
    elif len(str(row['Pair']).split('BTC')) > 1 and str(row['Pair']).split('BTC')[1] == '':
        if row['Side'] == 'SELL':
            received_cur += str(row['Pair']).split('BTC')[0]
            sent_cur += 'BTC'
        else:
            sent_cur += str(row['Pair']).split('BTC')[0]
            received_cur += 'BTC'
    elif len(str(row['Pair']).split('BNB')) > 1 and str(row['Pair']).split('BNB')[1] == '':
        if row['Side'] == 'SELL':
            received_cur += str(row['Pair']).split('BNB')[0]
            sent_cur += 'BNB'
        else:
            sent_cur += str(row['Pair']).split('BNB')[0]
            received_cur += 'BNB'
    elif len(str(row['Pair']).split('DAI')) > 1 and str(row['Pair']).split('DAI')[1] == '':
        if row['Side'] == 'SELL':
            received_cur += str(row['Pair']).split('DAI')[0]
            sent_cur += 'DAI'
        else:
            sent_cur += str(row['Pair']).split('DAI')[0]
            received_cur += 'DAI'
    elif len(str(row['Pair']).split('EUR')) > 1 and str(row['Pair']).split('EUR')[1] == '':
        if row['Side'] == 'SELL':
            received_cur += str(row['Pair']).split('EUR')[0]
            sent_cur += 'EUR'
        else:
            sent_cur += str(row['Pair']).split('EUR')[0]
            received_cur += 'EUR'
    else:
        print('Paire probleme :', row['Pair'])
        break
    if row['Side'] == 'SELL':
        received_quant += float(str(row['Executed']).split(f"{received_cur}")[0].replace(',',''))
        sent_quant += float(str(row['Amount']).split(f"{sent_cur}")[0].replace(',',''))
    else:
        sent_quant += float(str(row['Executed']).split(f"{sent_cur}")[0].replace(',',''))
        received_quant += float(str(row['Amount']).split(f"{received_cur}")[0].replace(',',''))
    if len(row['Fee'].split(f"{received_cur}")) > 1:
        fee_cur = received_cur
        fee_quant = row['Fee'].split(f"{received_cur}")[0]
    elif len(row['Fee'].split(f"{sent_cur}")) > 1:
        fee_cur = sent_cur
        fee_quant = row['Fee'].split(f"{sent_cur}")[0]
    elif len(row['Fee'].split('BNB')) > 1:
        fee_cur = 'BNB'
        fee_quant = row['Fee'].split('BNB')[0]
    else:
        print('Paire probleme :', row['Pair'])
        break
    new_datas['Sent Currency'].append(received_cur)        
    new_datas['Received Currency'].append(sent_cur)
    new_datas['Sent Quantity'].append(received_quant)        
    new_datas['Received Quantity'].append(sent_quant)
    new_datas['Fee Currency'].append(fee_cur)
    new_datas['Fee Quantity'].append(fee_quant)
    new_datas['Fees value in USDT'].append(0)
    new_datas['Trade value in USDT'].append(0)
    new_datas['Total trade value in USDT'].append(0)

new_datas = pandas.DataFrame(new_datas)
print(new_datas.head(10))
```

Get value in USDT

```python
trades = []
fees = []
total = []

for index, row in new_datas.iterrows():
    trade = 0.0
    fee = 0.0
    if row['Type'] == 'Deposit':
        if row['Received Currency'] != 'EUR':
            trade = get_value_in_USDT_width_binance(row['Timestamp'], row['Received Currency'], row['Received Quantity'])
        elif trade == 0.0:
            trade = get_value_in_USD(row['Date'], row['Received Quantity'])
    elif row['Type'] == 'Withdrawal':
        if row['Sent Currency'] != 'EUR':
            trade = get_value_in_USDT_width_binance(row['Timestamp'], row['Sent Currency'], row['Sent Quantity'])
        if trade == 0.0:
            trade = get_value_in_USD(row['Date'], row['Sent Quantity'])
    else:
        if row['Received Currency'] != 'EUR':
            trade = get_value_in_USDT_width_binance(row['Timestamp'], row['Received Currency'], row['Received Quantity'])
        if trade == 0.0:
            trade = get_value_in_USDT_width_binance(row['Timestamp'], row['Sent Currency'], row['Sent Quantity'])
    if row['Fee Currency'] != 0:
        if row['Fee Currency'] != 'EUR':
            fee = get_value_in_USDT_width_binance(row['Timestamp'], row['Fee Currency'], row['Fee Quantity'])
        # too many request
        # else:
            # fee = get_value_in_USD(row['Date'], row['Fee Quantity'])
    else:
        fee = 0
    trades.append(trade)
    fees.append(fee)
    total.append(trade + fee)


new_datas['Trade value in USDT'] = trades
new_datas['Fees value in USDT'] = fees
new_datas['Total trade value in USDT'] = total

new_datas.to_csv('binance_trades.csv')

print(new_datas)
```
